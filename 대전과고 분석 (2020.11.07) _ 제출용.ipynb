## 3. 새로운 컬럼생성 

```python
DSHS[['평균비행시간', '머신러닝시간', '연습시간', '노력시간']]
```

# 번아웃 학생이 각 컬럼에서 <br> 상위그룹에 속하는지 (75%), 하위그룹에 속하는지 (25%) 확인

```python
aim = DSHS[DSHS['번아웃']==1]
DSHS_predictors = DSHS.iloc[:, 1:]
criterion = DSHS_predictors.describe().loc[['25%', '75%'], :]
```

```python
columns = list(DSHS_predictors.columns)

good = []
worse = []

for col in columns:
    if aim[col].iloc[0] >= criterion[col][1]:
        good.append(col)
    elif aim[col].iloc[0] <= criterion[col][0]:
        worse.append(col)

len(good) + len(worse)
```

# 번아웃 후보 정하기

## 후보를 판단하기 위한 Judge 데이터프레임만들기

```python
a = DSHS[['혼자학습', '반복학습', '질문습관']] / 7    ## 7로 나눈 이유는 7이 '혼자학습', '반복학습', '질문습관' 에서 가장 높은 점수이기 때문
b = DSHS.loc[:, '국어학습태도':'생물학습태도']

hypothesis1 = pd.concat([a,b], axis = 1).sum(axis=1)

hypothesis1 = DSHS['gpa'] / hypothesis1
hypothesis1
```

```python
hypothesis2 = DSHS[['I유형.탐구형','과학분야_전문직','과학분야_숙련직',
                    '공학분야_전문직','공학분야_숙련직',
                    '과학학습선호','과학진로선호','이학공학연구',
                    '실험선호']]

hypothesis2 = (hypothesis2 - hypothesis2.min()) / (hypothesis2.max() - hypothesis2.min())

hypothesis2 = hypothesis2.sum(axis = 1)

hypothesis2
```

```python
DSHS['수평개인, 수직개인 평균'] = (DSHS['수평개인']+DSHS['수직개인'])/2

hypothesis3 = DSHS[['공동체의식','타인이해','인간관계','독립성','수평개인, 수직개인 평균']]

hypothesis3 = (hypothesis3 - hypothesis3.min()) / (hypothesis3.max() - hypothesis3.min())

hypothesis3['독립성'] = 1 - hypothesis3['독립성']
hypothesis3['수평개인, 수직개인 평균'] = 1 - hypothesis3['수평개인, 수직개인 평균']

hypothesis3 = hypothesis3.sum(axis = 1)
hypothesis3 
```

```python
Judge = pd.DataFrame({'가설 1': hypothesis1, '가설 2': hypothesis2, '가설 3': hypothesis3})
Judge
```

> ## 사분위를 이용하여 <br> Judge 데이터프레임을 1부터 4까지의 점수로 바꾸는 함수 만들기

```python
def classify(df):
    
    for column in df.columns:
        criteria = df[[column]].describe().loc[['25%', '50%', '75%'], :]

        Q1 = criteria[column][0]
        Q2 = criteria[column][1]
        Q3 = criteria[column][2]

        for i in range(len(df)):
            if (df[column][i] >= Q3):
                df[column][i] = 4
            elif (df[column][i] >= Q1):
                if (df[column][i] < Q2):
                    df[column][i] = 2
                else:
                    df[column][i] = 3
            else:
                df[column][i] = 1

classify(Judge)

Judge
```
