```
from sklearn.preprocessing import StandardScaler

stand_videos = []

stand = StandardScaler()

for video in user_3_list:
  video_stand = video.copy()
  video_stand[['Value']] = stand.fit_transform(video_stand[['Value']])
  stand_videos.append(video_stand)
  
user_3_list = stand_videos
user_3_list[0]
```

```
for i in range(len(resampled_videos)):
  x = resampled_videos[i]
  dummies = []
  for j in range(len(x)):
    if (x.index[j] - x.index[0]).seconds == 60:
      if (x.index[j] - x.index[0]).microseconds != 0:
        dummies.append(x.index[j])
    elif (x.index[j] - x.index[0]).seconds > 60:
      dummies.append(x.index[j])
    
    resampled_videos[i].drop(dummies, axis=0, inplace=True)
```
