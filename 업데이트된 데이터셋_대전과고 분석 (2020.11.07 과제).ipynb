```python
# 양의 값일수록 긍정적인 변화!!

gap1 = grade.iloc[:, 1] - grade.iloc[:, 0]
gap2 = grade.iloc[:, 2] - grade.iloc[:, 1]
gap3 = gap1 + gap2
gap4 = grade.iloc[:, 3] - grade.iloc[:, 0]
gap5 = grade.iloc[:, 3] - grade.iloc[:, 1]
gap6 = grade.iloc[:, 3] - grade.iloc[:, 2]

gap = pd.DataFrame({'1-2 difference': gap1, '2-3 difference': gap2, '1-3 difference': gap3, '1-total difference': gap4, 
                    '2-total difference': gap5, '3-total difference': gap6})

display(gap.describe())
gap.iloc[64, :]
```

```python
# 양의 값일수록 긍정적인 변화!!

rank_gap1 = grade_rank.iloc[:, 0] - grade_rank.iloc[:, 1]
rank_gap2 = grade_rank.iloc[:, 1] - grade_rank.iloc[:, 2]
rank_gap3 = rank_gap1 + rank_gap2
rank_gap4 = grade_rank.iloc[:, 0] - grade_rank.iloc[:, 3]
rank_gap5 = grade_rank.iloc[:, 1] - grade_rank.iloc[:, 3]
rank_gap6 = grade_rank.iloc[:, 2] - grade_rank.iloc[:, 3]

rank_gap = pd.DataFrame({'1-2 rank difference': rank_gap1, '2-3 rank difference': rank_gap2, '1-3 rank difference': rank_gap3,
                         '1-total rank difference': rank_gap4, '2-total rank difference': rank_gap5, 
                         '3-total rank difference': rank_gap6})

display(rank_gap.describe())
rank_gap.iloc[64, :]
```

```python
tendency_1_2 = ['Up' if diff >=0 else 'Down' for diff in rank_gap1]
tendency_2_3 = ['Up' if diff >=0 else 'Down' for diff in rank_gap2]

weight_score = []

for i in range(len(tendency_1_2)):
    if (tendency_1_2[i]=='Up'):
        if (tendency_2_3[i]=='Up'):
            weight_score.append(0.7 * 4/4)
        else:
            weight_score.append(0.7 * 2/4)
            
    else:
        if (tendency_2_3[i]=='Up'):
            weight_score.append(0.7 * 3/4)
        else:
            weight_score.append(0.7 * 1/4)

weight = pd.DataFrame({'1-2 tendency': tendency_1_2, '2-3 tendency': tendency_2_3, 'weight_score': weight_score})

gpa_with_weight = DSHS_clean['전체평점'] + weight_score
display(gpa_with_weight)
pd.DataFrame(gpa_with_weight).describe()
```

```python
weight_group = []

for i in range(len(tendency_1_2)):
    if (tendency_1_2[i]=='Up'):
        if (tendency_2_3[i]=='Up'):
            weight_group.append(4)
        else:
            weight_group.append(2)
            
    else:
        if (tendency_2_3[i]=='Up'):
            weight_group.append(3)
        else:
            weight_group.append(1)
            
weight_group = pd.DataFrame(weight_group, columns=['weight_group'])
weight_group
```

```python
first = []
second = []
third = []
fourth = []

for idx in range(len(weight_group)):
    if (weight_group.iloc[idx, :][0] == 1):
        first.append(idx)
    elif (weight_group.iloc[idx, :][0] == 2):
        second.append(idx)
    elif (weight_group.iloc[idx, :][0] == 3):
        third.append(idx)
    elif (weight_group.iloc[idx, :][0] == 4):
        fourth.append(idx)
        
print(first)
print(len(first))
print(second)
print(len(second))
print(third)
print(len(third))
print(fourth)
print(len(fourth))
```

```python
def classify(df):
    
    for column in df.columns:
        criteria = df[[column]].describe().loc[['25%', '50%', '75%'], :]

        Q1 = criteria[column][0]
        Q2 = criteria[column][1]
        Q3 = criteria[column][2]

        for i in range(len(df)):
            if (df[column][i] >= Q3):
                df[column][i] = 4
            elif (df[column][i] >= Q1):
                if (df[column][i] < Q2):
                    df[column][i] = 2
                else:
                    df[column][i] = 3
            else:
                df[column][i] = 1
         
classify(Judge)
```
